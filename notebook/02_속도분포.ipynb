# 02 · 속도 분포 확인 (보기용)

`04_speed.py`가 800개 세션에 속도를 재계산한 결과가 정상인지 그림으로 본다. 그림 **저장은 04_speed.py가** 하고(`python src/04_speed.py plots`), 이 노트북은 그 함수를 불러 **화면으로 보기만** 한다.

**데이터 흐름**
- **입력**: `data/processed/speed_summary.csv` — 04_speed.py가 800개 세션 각각의 속도를 재계산해 요약한 표(세션당 한 줄)
- **그 표 내용**: 이동속도 중앙/p95/max, 정지(<5km/h) 프레임 비율, 속도 결측률
- **목적**: 한 샘플이 아니라 **전 세션에서 속도가 사리에 맞나** 확인 (이동속도 ~30, 정지비율, 교차로 편차)

```python
import importlib.util
import pandas as pd

# 04_speed.py의 함수 불러오기 (숫자로 시작하는 파일명이라 importlib 사용)
spec = importlib.util.spec_from_file_location('speed', '../src/04_speed.py')
speed = importlib.util.module_from_spec(spec)
spec.loader.exec_module(speed)

s = pd.read_csv('../data/processed/speed_summary.csv')
print(f'세션 {len(s)}개')
s.head()
```

## 1) 세션별 이동 속도 분포 — ~30km/h 중심이면 정상

```python
speed.fig_moving_speed(s)
```

## 2) 정지(<5km/h) 프레임 비율 분포 — 신호 대기가 많아 절반 이상

```python
speed.fig_stopped(s)
```

## 3) 교차로별 이동 속도 — 속도 환경 편차(전이성 분석의 covariate)

```python
speed.fig_speed_by_intersection(s)
```
